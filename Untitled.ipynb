```python
def get_valid_box_on_screen(img , box_width , box_height):
    im_size = img.shape
    im_w = im_size[1]
    im_h = im_size[0]
    w_half = im_w/2
    h_half = im_h/2
    
    w_left = (int)(w_half - (box_width/2))
    h_up = (int)(h_half - (box_height/2))
    
    cv2.rectangle(img , (w_left,h_up), (w_left + box_width, h_up + box_height), (0,255,0), 2) #making rectangle around detected face
    return img
```
